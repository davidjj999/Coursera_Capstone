```python
#Import the required libraries for this project
import pandas
import numpy
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
#Import the source data
concrete_data=pandas.read_csv('https://cocl.us/concrete_data')
```

```python
#Verify the input data frame looks as expected
concrete_data.head()
```

```python
#Predictors are all columns except strength
predictors=concrete_data[concrete_data.columns[concrete_data.columns!='Strength']]
```

```python
#Target is strength
target=concrete_data['Strength']
```

```python
#Verify target array shape
target.head()
```

```python
#Verify predictors data frame no longer includes strength
predictors.head()
```

```python
#Scale input and target
scaled_predictors=(predictors-predictors.mean())/predictors.std()
scaled_target=(target-target.mean())/target.std()
```

```python
#Use number of predictor columns for model
ncols=scaled_predictors.shape[1]
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Define the shape of the neural network.  10 nodes in hiden layer.
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(ncols,)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['mse'])
```

```python
### Part A ###
#Uses non-normalized data

#Repeat this 50 times:
#1. split data frame into 70/30 train/test
#2. fit the model with the new training set
#3. generate predicted Y values using test set
#4. compute MSE from test set values
#5. append MSE to list
errors=[]
for i in range(50):
    x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
    model.fit(x_train, y_train, epochs=50)
    prediction=model.predict(x_test)
    errors.append(mean_squared_error(y_test,prediction))
```

```python
import numpy
output=numpy.array(errors)
```

```python
output.mean()
```

```python
output.std()
```

```python
### Part B ###
#Uses normalized data

#Repeat this 50 times:
#1. split data frame into 70/30 train/test
#2. fit the model with the new training set
#3. generate predicted Y values using test set
#4. compute MSE from test set values
#5. append MSE to list
norm_errors=[]
for i in range(50):
    x_train, x_test, y_train, y_test = train_test_split(scaled_predictors, scaled_target, test_size=0.3)
    model.fit(x_train, y_train, epochs=50)
    prediction=model.predict(x_test)
    norm_errors.append(mean_squared_error(y_test,prediction))
```

```python
norm_output=numpy.array(norm_errors)
```

```python
norm_output.mean()
```

```python
norm_output.std()
```

```python
#mean is very diffferent than part A due to normalized data
norm_output.mean()-output.mean()
```

```python
### Part C ###
#Uses normalized data, 100 epochs

#Repeat this 50 times:
#1. split data frame into 70/30 train/test
#2. fit the model with the new training set
#3. generate predicted Y values using test set
#4. compute MSE from test set values
#5. append MSE to list
epoch_errors=[]
for i in range(50):
    x_train, x_test, y_train, y_test = train_test_split(scaled_predictors, scaled_target, test_size=0.3)
    model.fit(x_train, y_train, epochs=100)
    prediction=model.predict(x_test)
    epoch_errors.append(mean_squared_error(y_test,prediction))
```

```python
epoch_output=numpy.array(epoch_errors)
```

```python
epoch_output.mean()
```

```python
epoch_output.std()
```

```python
#Average MSE for part C is smaller than part B
epoch_output.mean()-norm_output.mean()
```

```python
### Part D ###
#Uses normalized data, three hidden layers
#Define the shape of the neural network.  10 nodes in hiden layer.
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(ncols,)))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['mse'])
```

```python
#Repeat this 50 times:
#1. split data frame into 70/30 train/test
#2. fit the model with the new training set
#3. generate predicted Y values using test set
#4. compute MSE from test set values
#5. append MSE to list
hidden_errors=[]
for i in range(50):
    x_train, x_test, y_train, y_test = train_test_split(scaled_predictors, scaled_target, test_size=0.3)
    model.fit(x_train, y_train, epochs=50)
    prediction=model.predict(x_test)
    hidden_errors.append(mean_squared_error(y_test,prediction))
```

```python
hidden_output=numpy.array(hidden_errors)
```

```python
hidden_output.mean()
```

```python
hidden_output.mean()
```

```python
#Average MSE with more layers is smaller than part B
hidden_output.mean()-norm_output.mean()
```
